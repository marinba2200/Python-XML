```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head()
```

```python
df_state = df.groupby( "State" ).count() #計算次數用.count() #依項目集合值用.groupby(項目)
```

```python
df_state
```

```python
df_state.sort_values( by = "Time" , ascending = False ) #排序，ascending是否由小到大，產生新的dataframe(預設是不會覆蓋原始資料表)
```

```python
df_state.sort_values( by = "Time" , ascending = False , inplace = True ) #排序，ascending是否由小到大，覆蓋原始資料表
```

```python
df_state.head(10)
```

```python
df_state[ : 10 ].Time.plot() #預設折線圖 (:10切片前10)
```

```python
df_state[ : 10 ].Time.plot( kind = "bar" ) #長條圖
```
